```python
import numpy as np

def diagonalDom(x):
    size = x.shape
    for i in range(0, size[0]):
        sum = 0
        for j in range(0, size[1]):
            sum += abs(x[i][j])
        if (x[i][i] < (sum - abs(x[i][i]))):
            print(x)
            return False
    print(x)
    return True

x = np.array([[3,1],[1,5]])
diagonalDom(x)
```
