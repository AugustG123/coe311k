```python
import numpy as np

import matplotlib.pyplot as plt

from scipy.interpolate import CubicSpline

 

# Define known data points

x = np.array([0, 1, 2, 3, 4, 5])

y = np.array([0, 1, 0, 1, 0, 1])

# Create a cubic spline interpolation of the data

cs = CubicSpline(x, y, bc_type='natural')# Generate new x values for a smooth plot

x_interp = np.linspace(0, 5, 100)

y_interp = cs(x_interp)

poly = np.polyfit(x,y,3)

y_interp2 = np.polyval(poly, x_interp)

plt.plot(x_interp, y_interp, color = 'blue')
plt.plot(x_interp, y_interp2, color = 'red')

#Cubic spline is much better because the data oscilates. Degree choice is a little hard 
#here because finite degree polynomials cannot oscilate forever but I dont want to blanatly overfit as well
```
