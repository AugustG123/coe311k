```python
import random
import math

def approxPi(n):
    count = 0
    for i in range(n):
        x = random.random();
        y = random.random();
        if (x*x + y*y) < 1:
            count += 1
    return 4*count/n

values = []
n = 10
values.append(approxPi(n))
n = 100
values.append(approxPi(n))
n = 1000
values.append(approxPi(n))
n = 10000
values.append(approxPi(n))

meanError = []
for value in values:
    meanError.append(abs(value - 3.14))
print(meanError)
```
