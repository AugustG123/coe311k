```python
import math
primes = []
for i in range(3,1000):
    prime = True
    for j in range(2,i):
        if i % j == 0:
            prime = False
    if prime:
        primes.append(i)
        
print(primes)
```

```python
correctValues = []
for a in range(1,100):
    for b in range(1,100):
        x = math.sqrt(a*a + b*b)
        if int(x) == x:
            correctValues.append(str(a) + ", " + str(b))
```

```python
print(correctValues)
```

```python
import random
import numpy as np
n = 1

def approxPi(n):
    count = 0
    for i in range(n):
        x = random.random();
        y = random.random();
        if (x*x + y*y) < 1:
            count += 1
    return 4*count/n

n = np.linspace(10,1000,49.5)
print(n)
```
