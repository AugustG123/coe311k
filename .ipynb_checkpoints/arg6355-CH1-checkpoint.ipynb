```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10,10)

plt.subplot(2,2,1)
plt.hexbin(x, np.arcsinh(x), gridsize = 20)
plt.subplot(2,2,2)
plt.scatter(x, np.arcsinh(x))
plt.plot()

plt.subplot(223)
plt.stackplot(x,np.arcsinh(x))
plt.plot()

plt.subplot(224)
plt.hist2d(x,np.arcsinh(x))
plt.plot()
```
