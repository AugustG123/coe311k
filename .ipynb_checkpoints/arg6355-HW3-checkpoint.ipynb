```python
import math
import numpy as np
import matplotlib.pyplot as plt

def taylorSin(x,n):
    value = 0
    for i in range(0,n):
        value += (((-1)**i)*(x**(2*i+1))) / (math.factorial(2*i+1))
    return value

x = np.arange(0, 10.0001, 0.1)

t1 = []
t2 = []
t3 = []
t4 = []

for value in x:
    t1.append(abs(np.sin(value) - taylorSin(value, 1)))
    t2.append(abs(np.sin(value) - taylorSin(value, 2)))
    t3.append(abs(np.sin(value) - taylorSin(value, 3)))
    t4.append(abs(np.sin(value) - taylorSin(value, 4)))
    
plt.plot(x,t1,color='blue')
plt.plot(x,t2,color='red')
plt.plot(x,t3,color='green')
plt.plot(x,t4,color='yellow')

plt.legend(["n=1", "n=2", "n=3", "n=4"])
plt.xlabel("x")
plt.ylabel("Log of Absolute Error")
plt.yscale("log")
plt.title("Taylor Approximation Error for sin(x)")
plt.show()

#The Truncation Error near 0 is smaller for more terms, which is to be expected from the definition of a maclaurin series. 
#However the truncation error increases for more terms as the x gets farther away from 0. 
#This is because a polynomial(with finite terms) cant oscillate. The error of a taylor series is less than the absolute
#value of the next term(because of lagrange error bound), since more terms means the next term is higher power the error
#can be higher when x is large enough.
```
