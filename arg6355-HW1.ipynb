```python
import random
import numpy as np
import matplotlib.pyplot as plt
n = 1

def approxPi(n):
    count = 0
    for i in range(n):
        x = random.random();
        y = random.random();
        if (x*x + y*y) < 1:
            count += 1
    return 4*count/n
n = []
for i in range(10,1000,20):
    n.append(i)
error = []
for value in n:
    error.append(abs(approxPi(value)- np.pi))
plt.plot(n,error)
```
