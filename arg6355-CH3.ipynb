```python
import random
import numpy as np
import matplotlib.pyplot as plt
n = 1

def approxPi(n):
    count = 0
    for i in range(n):
        x = random.random();
        y = random.random();
        if (x*x + y*y) < 1:
            count += 1
    return 4*count/n

errorAbs = []
errorRel = []
for i in range(0,100):
    x = approxPi(1000)
    errorAbs.append(abs(x- np.pi))
    errorRel.append(abs(x- np.pi)/np.pi)
    
plt.plot(errorAbs, label='Absolute Error')
plt.plot(errorRel, label='Relative Error')
plt.ylabel('Error')
plt.xlabel('Case')
plt.show()
```
