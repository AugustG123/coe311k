```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10,10)

plt.subplot(221)
plt.hexbin(x, np.arcsinh(x), gridsize = 20)
plt.title("plot 1")

plt.subplot(222)
plt.scatter(x, np.arccosh(x))
plt.plot()
plt.title("plot 2")

plt.subplot(223)
plt.stackplot(x,np.tanh(x))
plt.plot()
plt.title("plot 3")

plt.subplot(224)
plt.hist2d(x,np.sinh(x))
plt.plot()
plt.title("plot 4")

plt.subplots_adjust(left = 0, right = .7, bottom = 0, top = .7, wspace = .7, hspace = .7)
```
