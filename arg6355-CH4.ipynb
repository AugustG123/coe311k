```python
import matplotlib.pyplot as plt
import numpy as np
def piApproxTerm(n):
    return (4*(-1)**n)/(2*n+1)
value = 0
error = []
x = []
for i in range(0, 5001):
    value += piApproxTerm(i)
    if i % 100 == 0:
        error.append(abs((value - np.pi)/np.pi))
plt.plot(error)
plt.ylabel('Error')
plt.show()
```
